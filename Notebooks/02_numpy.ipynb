Créez un array bidimensionnel avec des valeurs quelconques, calculer le cube de cet array


```python
# on peut générer un array à partir d'une liste
arr4=np.array([[3,4,5],[4,3,6]])
print(arr4)
```

Construire une matrice rectangulaire de taille 3 x 2 avec une suite de nombres commençant à 5, calculer le produit de la transposée par la matrice.


```python
arr8=np.arange(5,11).reshape(3,2)
```

```python
print(np.dot(arr8.T,arr8))
```

```python
print(arr8.T.dot(arr8))
```

Générer un array de nombres aléatoires puis remplacer les valeurs plus petites que 0 par 0. Trier l’array obtenu


```python
arr9=np.random.randn(1000000)
```

```python
%%timeit
# mauvaise approche
for i,val in enumerate(arr9):
    if val<0:
        arr9[i]=0
```

```python
%%timeit 
# bonne approche
arr10=np.where(arr9<0,0,arr9)
```

```python
print(np.sort(np.where(arr9<0,0,arr9)))
```

Calculer des statistiques sur les colonnes d'une matrice

```python
arr_stat=np.random.randn(10000).reshape(1000,10)
```

```python
print(arr_stat.mean(axis=0))
```

```python
print(np.mean(arr_stat,axis=0))
```

Générer un array de nombres aléatoires tirés d’une loi normale et en calculer l’exponentielle


```python
print(np.exp(arr_stat))
```
